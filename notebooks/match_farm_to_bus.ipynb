```python
import numpy as np
import pandas as pd
```

```python
wind_data = pd.read_csv('../data/raw/wind_cap_loc.csv')
wind_data = wind_data.set_index("Ref ID")

grid_data_bus = pd.read_csv('../data/raw/network/substations_geo.csv')
```

```python
def get_lat_lon(ref_id):
    try:
        row = wind_data.loc[ref_id]
        return row["Latitude"], row["Longitude"]
    except KeyError:
        raise ValueError(f"Ref ID {ref_id} not found")
```

```python
def get_distance (lat1, long1, lat2, long2):
    dlat = lat1 - lat2
    dlong = long1 - long2
    distance = np.sqrt(dlat**2 + dlong**2)
    return distance
```

```python
def get_nearest_bus(ref_id):
    wind_lat, wind_long = get_lat_lon(ref_id)

    distances = get_distance(
        wind_long,
        wind_lat,
        grid_data_bus["longitude"],
        grid_data_bus["latitude"]
    )

    i = distances.idxmin()  

    nearest = grid_data_bus.loc[i]
    return nearest
```

```python
print(get_nearest_bus(2477))
```
